```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
file_path = "deliveries_updated_mens_ipl_upto_2024 - deliveries_updated_mens_ipl_upto_2024.csv"
df = pd.read_csv(file_path)

# Display shape (rows, columns) and first 5 rows
print("Dataset Shape:", df.shape)
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python

categorical_cols = ['batting_team', 'bowling_team', 'batsman', 'non_striker', 'bowler', 'dismissal_kind', 'player_dismissed']
for col in categorical_cols:
    df[col] = df[col].astype('category')

df['date'] = pd.to_datetime(df['date'], errors='coerce')

numeric_cols = ['over', 'ball', 'batsman_runs', 'extras', 'isWide', 'isNoBall', 'Byes', 'LegByes', 'Penalty']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df = df.drop_duplicates()
df.info()
```

organised data into categories and cleaned it



```python

total_matches = df['matchId'].nunique()
print("Total matches:", total_matches)


teams = df['batting_team'].unique()
print("Total teams:", len(teams))
print("Teams:", teams)


unique_batsmen = df['batsman'].nunique()
unique_bowlers = df['bowler'].nunique()
print("Unique batsmen:", unique_batsmen)
print("Unique bowlers:", unique_bowlers)


total_runs = df['batsman_runs'].sum() + df['extras'].sum()
print("Total runs scored:", total_runs)


print("\nOvers summary:")
print(df['over'].describe())
```

```python


#  Runs per Over
runs_per_over = df.groupby('over')['batsman_runs'].sum() + df.groupby('over')['extras'].sum()

balls_per_over = df.groupby('over')['ball'].count()

avg_runs_per_over = (runs_per_over / balls_per_over) * 6
avg_runs_per_over = avg_runs_per_over.reset_index()  # convert to DataFrame

fig = px.line(
    avg_runs_per_over,
    x='over',
    y=0,
    title="Average Runs per Over (All Matches)",
    markers=True
)

fig.show()

```

```python


df['year'] = df['date'].dt.year

# Identify boundaries for milestone calculation.create is_four and is_six.
df['is_four'] = (df['batsman_runs'] == 4).astype(int)
df['is_six'] = (df['batsman_runs'] == 6).astype(int)


# =====================
# 1. Top 5 Batsmen by Runs (All-Time)
# =====================

# All-time
batsman_runs = (
    df.groupby('batsman')['batsman_runs']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

fig1 = px.bar(
    batsman_runs, x='batsman_runs', y='batsman',
    orientation='h', title="Top 5 Batsmen by Runs (All-Time)",
    text='batsman_runs', color='batsman_runs',
    color_continuous_scale="Viridis"
)
fig1.update_layout(yaxis=dict(autorange="reversed"))
fig1.show()
```

```python
# =====================
# 2. Top 5 Batsmen by Strike Rate (min 500 balls)
# =====================

batsman_stats = (
    df.groupby('batsman')
    .agg({'batsman_runs':'sum','ball':'count'})
    .reset_index()
)

batsman_stats['strike_rate'] = (batsman_stats['batsman_runs'] / batsman_stats['ball']) * 100
qualified = batsman_stats[batsman_stats['ball'] >= 500]

top5_sr = qualified.sort_values('strike_rate', ascending=False).head(5)

fig3 = px.bar(
    top5_sr, x='strike_rate', y='batsman',
    orientation='h', title="Top 5 Batsmen by Strike Rate (min 500 balls)",
    text=top5_sr['strike_rate'].round(2), color='strike_rate',
    color_continuous_scale="Plasma"
)
fig3.update_layout(yaxis=dict(autorange="reversed"))
fig3.show()
```

 Identifying Elite "Finishers": These players are prime candidates to be designated as "finishers." Their primary role is to bat in the final overs of an innings (e.g., overs 16-20) to maximize the score. A team captain can hold these batsmen back, preserving their wickets for a final, decisive assault when the run rate needs a dramatic boost.
 Defining the "Impact Player" Role: In tournaments with substitution rules (like the IPL's "Impact Player"), these are perfect candidates for that role. A team can bring one of these batsmen in late in the innings purely to score quickly for 2-3 overs, providing a massive tactical advantage without having to field them for the entire match.

```python
# =====================
# 3. Boundary Hitters  Most Sixes + Fours
# =====================

boundary_stats = (
    df.groupby('batsman')[['is_four','is_six']]
    .sum()
    .sort_values('is_six', ascending=False)
    .head(5)
    .reset_index()
)

fig4 = px.bar(
    boundary_stats, x=['is_four','is_six'], y='batsman',
    orientation='h', title="Top 5 Boundary Hitters (Fours & Sixes)",
    barmode='group'
)
fig4.update_layout(yaxis=dict(autorange="reversed"), xaxis_title="Count")
fig4.show()
```

```python

# =====================
# 4. Phase-wise Batting (Powerplay, Middle, Death)
# =====================

def phase(row):
    if 1 <= row['over'] <= 6:
        return "Powerplay (1-6)"
    elif 7 <= row['over'] <= 15:
        return "Middle (7-15)"
    else:
        return "Death (16-20)"

df['phase'] = df.apply(phase, axis=1)

phase_runs = (
    df.groupby(['phase','batsman'])['batsman_runs']
    .sum()
    .reset_index()
)

top5_phase = (
    phase_runs
    .sort_values(['phase','batsman_runs'], ascending=[True,False])
    .groupby('phase').head(5)
)

fig5 = px.bar(
    top5_phase, x='batsman_runs', y='batsman',
    color='phase', orientation='h',
    title="Top 5 Batsmen by Runs in Each Phase"
)
fig5.update_layout(yaxis=dict(autorange="reversed"))
fig5.show()
```

Powerplay (Overs 1-6)
This data shows who can best exploit the early field restrictions. It helps teams select an optimal opening pairpairing an aggressor with a stable anchor. Opponents can identify their best opening bowlers to either take crucial early wickets or contain the initial run-scoring surge.

Middle Overs (Overs 7-15)
This phase highlights players who prevent the innings from stalling. Teams can identify batsmen who are experts at rotating the strike and handling spin. Bowling teams, in turn, know when to deploy their best spinners to control the run rate and build pressure.
Death Overs (Overs 16-20)
This data pinpoints the most valuable players when the match is on the line. It identifies elite "finishers" who dramatically accelerate the score. For the opposition, it reveals which of their bowlers are reliable "death specialists" who can execute plans under pressure to close out the innings.

```python

# =====================
# 6. Milestones  50s and 100s per batsman
# =====================

# Runs per batsman per match
match_runs = (
    df.groupby(['matchId','batsman'])['batsman_runs']
    .sum()
    .reset_index()
)

milestones = (
    match_runs.groupby('batsman')
    .agg(
        fifties = ('batsman_runs', lambda x: (x.between(50,99)).sum()),
        hundreds = ('batsman_runs', lambda x: (x >= 100).sum())
    )
    .reset_index()
    .sort_values('hundreds', ascending=False)
    .head(5)
)

fig7 = px.bar(
    milestones, x=['fifties','hundreds'], y='batsman',
    orientation='h', barmode='group',
    title="Milestones: 50s & 100s by Batsmen"
)
fig7.update_layout(yaxis=dict(autorange="reversed"))
fig7.show()
```

```python
import pandas as pd
import plotly.express as px

# Assuming 'df' is your DataFrame

# PREP: create total_run, dot-ball, and is_wicket indicators
# =====================
df['total_run'] = df['batsman_runs'] + df['extras']
df['is_dot'] = (df['total_run'] == 0).astype(int)
# FIX: Add this line to create the 'is_wicket' column
df['is_wicket'] = (df['dismissal_kind'].notna()).astype(int)


# =====================
# 1. Top 5 Bowlers by Wickets (All-Time)
# =====================
bowler_wickets = (
    df.groupby('bowler')['is_wicket']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

fig1 = px.bar(
    bowler_wickets, x='is_wicket', y='bowler',
    orientation='h', text='is_wicket',
    title="Top 5 Bowlers by Wickets (All-Time)",
    color='is_wicket', color_continuous_scale="Viridis"
)
fig1.update_layout(yaxis=dict(autorange="reversed"))
fig1.show()
```

```python
# =====================
# 2. Best Economy (min 100 overs = 600 balls)
# =====================
bowler_stats = (
    df.groupby('bowler')
    .agg({'total_run':'sum','ball':'count'})
    .reset_index()
)
bowler_stats['overs'] = bowler_stats['ball'] / 6
bowler_stats['economy'] = bowler_stats['total_run'] / bowler_stats['overs']

qualified = bowler_stats[bowler_stats['overs'] >= 100]
top5_economy = qualified.sort_values('economy').head(5)

fig2 = px.bar(
    top5_economy, x='economy', y='bowler',
    orientation='h', text=top5_economy['economy'].round(2),
    title="Top 5 Economical Bowlers (min 100 overs)",
    color='economy', color_continuous_scale="Plasma"
)
fig2.update_layout(yaxis=dict(autorange="reversed"))
fig2.show()
```

```python
# =====================
# 4. Maiden Overs
# =====================
df['over_id'] = df['matchId'].astype(str) + "_" + df['inning'].astype(str) + "_" + df['over'].astype(str)

maiden_check = (
    df.groupby(['bowler','over_id'])['total_run']
    .sum()
    .reset_index()
)
maiden_check['is_maiden'] = (maiden_check['total_run'] == 0).astype(int)

maiden_count = (
    maiden_check.groupby('bowler')['is_maiden']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

fig3 = px.bar(
    maiden_count, x='is_maiden', y='bowler',
    orientation='h', text='is_maiden',
    title="Top 5 Bowlers by Maiden Overs",
    color='is_maiden', color_continuous_scale="Cividis"
)
fig3.update_layout(yaxis=dict(autorange="reversed"))
fig3.show()
```

```python

# =====================
# 5. Phase-wise Bowling (Economy + Wickets)
# =====================
def bowling_phase(row):
    if 1 <= row['over'] <= 6:
        return "Powerplay (1-6)"
    elif 7 <= row['over'] <= 15:
        return "Middle (7-15)"
    else:
        return "Death (16-20)"

df['bowling_phase'] = df.apply(bowling_phase, axis=1)

phase_stats = (
    df.groupby(['bowler','bowling_phase'])
    .agg({'is_wicket':'sum','total_run':'sum','ball':'count'})
    .reset_index()
)
phase_stats['overs'] = phase_stats['ball'] / 6
phase_stats['economy'] = phase_stats['total_run'] / phase_stats['overs']

top_phase = (
    phase_stats
    .sort_values(['bowling_phase','is_wicket'], ascending=[True,False])
    .groupby('bowling_phase').head(5)
)

fig4 = px.scatter(
    top_phase, x='economy', y='is_wicket',
    size='is_wicket', color='bowling_phase', hover_name='bowler',
    title="Bowler Performance in Phases (Wickets vs Economy)"
)
fig4.show()
```

A glance at the plot shows which of the three has the best record of either taking wickets or controlling runs during this specific phase. This allows the captain to make a tactical decision based on data, not just gut feeling.

```python

# =====================
# 6. Bowler vs Specific Team
# =====================
team = "Mumbai Indians"

bowler_vs_team = (
    df[df['batting_team'] == team]
    .groupby('bowler')['is_wicket']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

fig5 = px.bar(
    bowler_vs_team, x='is_wicket', y='bowler',
    orientation='h', text='is_wicket',
    title=f"Top Bowlers vs {team}",
    color='is_wicket', color_continuous_scale="Magma"
)
fig5.update_layout(yaxis=dict(autorange="reversed"))
fig5.show()
```

 Exploit Favorable Matchups: If a bowler has an excellent historical record (low economy, high wickets) against a certain team, they are a "matchup weapon." The insight is to ensure they are in the playing XI for that fixture and to use them against the opposition's key batsmen to maximize their psychological and statistical advantage.

 Avoid or Hide Unfavorable Matchups: If the data shows a bowler consistently performs poorly against a specific team (e.g., their star batsman plays them very well), the insight is to minimize their exposure. The captain can choose to rest that bowler for the match or use them in less critical phases of the game to protect them and prevent the opposition from targeting a known weakness.

 Opposition's Game Plan: This data also reveals your opponent's strategy. If your bowler has a poor record, you can anticipate that the opposition batsmen will be instructed to attack them. This allows your captain to set defensive fields or have a backup plan ready if that bowler is targeted.

```python


# =====================
# FIX 1: Create a total runs column first
# =====================
df['total_runs'] = df['batsman_runs'] + df['extras']

# =====================
# 1) Win probability: Batting First vs Chasing
# =====================

# FIX 2: Group by the new 'total_runs' column
inning_totals = df.groupby(['matchId','inning','batting_team'])['total_runs'].sum().reset_index()

# Split into innings 1 and 2
inning1 = inning_totals[inning_totals['inning']==1].rename(columns={'batting_team':'team1','total_runs':'score1'})
inning2 = inning_totals[inning_totals['inning']==2].rename(columns={'batting_team':'team2','total_runs':'score2'})

# Merge to create match summary
match_summary = inning1.merge(inning2, on='matchId')

# Winner column (this logic is now correct because the scores are correct)
match_summary['winner'] = match_summary.apply(
    lambda x: x['team1'] if x['score1']>x['score2'] else x['team2'],
    axis=1
)

# Tag batting first vs chaser
match_summary['bat_first'] = match_summary['team1']
match_summary['chaser'] = match_summary['team2']

# Calculate win percentages
bat_first_wins = (match_summary['winner'] == match_summary['bat_first']).mean()*100
chasing_wins = (match_summary['winner'] == match_summary['chaser']).mean()*100

# Pie chart
fig1 = px.pie(
    names=['Batting First Wins','Chasing Wins'],
    values=[bat_first_wins,chasing_wins],
    title="Win Probability: Batting First vs Chasing"
)
fig1.show()
```

This insight directly informs the captain's decision at the coin toss. It reveals whether setting a target or chasing is statistically more likely to lead to a win, guiding the crucial choice to bat or bowl first.

```python
# 3) Dismissals by type
dismissals = (
    df[df['dismissal_kind'].notnull()]
    .groupby('dismissal_kind')['player_dismissed']
    .count().reset_index()
    .sort_values('player_dismissed',ascending=False)
)

fig3 = px.bar(
    dismissals, x='player_dismissed', y='dismissal_kind',
    orientation='h', text='player_dismissed',
    title="Dismissals by Type",
    color='player_dismissed', color_continuous_scale="Plasma"
)
fig3.update_layout(yaxis=dict(autorange="reversed"))
fig3.show()
```

```python
# Filter for Abhishek Sharma (last 4 years)
abhishek = df[(df['batsman'] == "Abhishek Sharma") & (df['year'] >= df['year'].max() - 3)]

abhishek_dismissals = (
    abhishek[abhishek['dismissal_kind'].notna()]
    .groupby('dismissal_kind')
    .size()
    .reset_index(name='count')
)

fig1 = px.pie(
    abhishek_dismissals,
    values='count',
    names='dismissal_kind',
    title="Dismissal Types of Abhishek Sharma (Last 4 Years)",
    hole=0.3  # donut-style for better readability
)
fig1.show()
```

 For the Opposition Captain: The data shows the highest percentage method to get him out. If he is frequently dismissed caught at deep mid-wicket, the captain can place a fielder there and bring on an off-spinner. If he is prone to being bowled or LBW by fast inswingers, the plan is to attack him with that type of bowling during the Powerplay.

 For His Own Coach: This analysis pinpoints technical flaws to address in training. A high number of stumped dismissals points to a need for better footwork against spin, while frequent caught dismissals in the cover region suggest a need to work on shot selection for his drives.
